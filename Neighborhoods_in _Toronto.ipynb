```python
import numpy as np 
import pandas as pd
```

```python
# define the dataframe columns
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood

```python
df.columns=['Postcode','Borough','Neighbourhood']

df.drop([0],axis=0,inplace=True)

df.reset_index()
```

 Ignore cells with a borough that is Not assigned.

```python
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
```

combined the two nigeighborhood with the same postal address into one row with the neighborhoods 

```python
df1=df.groupby("Postcode").agg(lambda x:','.join(set(x)))
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df1.loc[df1['Neighbourhood']=="Not assigned",'Neighbourhood']=df1.loc[df1['Neighbourhood']=="Not assigned",'Borough']
```

Now add the columns 

```python
df_csv = pd.read_csv('http://cocl.us/Geospatial_data')
df_csv.head(5)
df_csv.columns
```

```python
df_csv.columns=['Postcode','Latitude','Longitude']

df_csv.drop([0],axis=0,inplace=True)

df_csv.reset_index()

```

```python
df_csv.rename(index=str, columns={"Postal Code": "Postcode"})
```

```python
result = pd.merge(df1,
                 df_csv[['Postcode', 'Latitude', 'Longitude']], on="Postcode")
result.head(11)
```
